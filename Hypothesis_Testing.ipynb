**Q1.** What is a null hypothesis (H₀) and why is it important in hypothesis testing?

**Ans-** **NULL HYPOTHESIS (H₀):** It is a statement that assumes **no effect**, **no difference**, or **no relationship** exists in the population. It is the hypothesis that is tested using sample data.

**Example**

Suppose a company claims that the average battery life of its product is **10 hours**.

- **Null Hypothesis (H₀):** μ = 10 hours

  (The average battery life is 10 hours.)
- **Alternative Hypothesis (H₁):** μ ≠ 10 hours

  (The average battery life is different from 10 hours.)

  ## **Why it is important in hypothesis testing?**

  **1.** It provides a baseline for comparison in hypothesis testing.

  **2.** It helps determine whether observed difference are statistically significant or due to random chance.
  
  **3.** It form the basis for calculating test statistics & p-values.

  **4.** Guides decision-making by allowing us to either:

  - **Reject H₀** when there is sufficient evidence against it.
  - **Fail to reject H₀** when there is insufficient evidence.


**Q2.**  What does the significance level (α) represent in hypothesis testing?

**Ans-** The **significance level (α)** defines how much risk we are willing to take of making a wrong decision by rejecting the null hypothesis when it is actually true.

  **Common Values of α**

 -  **α** = **0.05** (5%) – most commonly used

 - **α** = **0.01** (1%)

 - **α** = **0.10** (10%)

 **Importance of α**

 **1.** Sets the criterion for rejecting H₀.

 **2.** Controls the risk of Type I error.

 **3.** Determines statistical significance:
   - If **p-value <= α**, reject H₀.
   - If **p-value > α**, fail to reject H₀

   **Example**

   Supoose:  **α** = 0.05
   
  - If the p-value is **0.03**, since **0.03 < 0.05**, reject H₀.
  - If the p-value is **0.08**, since **0.08 > 0.05**, fail to reject H₀.


            





**Q3.** Differentiate between Type I and Type II errors.

**Ans-**

   **Type I Error:** Saying **"Yes"** when the answer is actually **"No"**.

   **Type II Error:** Saying **"No"** when the answer is actually **"Yes"**.

| **Type I Error (False Positive)**                       | **Type II Error (False Negative)**                                        |
| ------------------------------------------------------- | ----------------------------------------------------------|
| Rejecting a true null hypothesis ((H_0)).               | Failing to reject a false null hypothesis ((H_0)).                            |
| Also called a **false positive**.                       | Also called a **false negative**.                                       |
| Occurs with probability **α (significance level)**.     | Occurs with probability **β**.                                        |
| Concluding there is an effect when none exists.         | Concluding there is no effect when one actually exists.                       |
| **Example:** Declaring a new drug effective when it is not. | **Example:** Concluding a new drug is ineffective when it actually works.  |


**Q4.** Explain the difference between a one-tailed and two-tailed test. Give an example of each.

**Ans-**

| **One-Tailed Test**                                                         | **Two-Tailed Test**                                                     |
| --------------------------------------------------------------------------- | ----------------------------------------------------------------------- |
| Tests for a specific direction (greater than, less than, increase & decrease). | Tests for any difference, regardless of direction.                      |
| The rejection region is located in only one tail of the distribution.       | The rejection regions are split between both tails of the distribution. |
| Alternative hypothesis uses (>) or (<).                                  | Alternative hypothesis uses (!=).                                     |
| More powerful for detecting an effect in one direction.                     | Detects effects in both directions.                                     |

**Example of a One-Tailed Test**

A company claims that a new battery lasts more than **10 hours**.

- **Null Hypothesis (H0):** μ = 10 hours
- **Alternative Hypothesis (Ha):** = μ > 10 hours

  This is a right-tailed test because we are interested only in whether the average battery life is greater than 10 hours.

**Example of a Two-Tailed Test**

 A manufacturer claims that the average weight of a cereal box is **500 g**.

- **Null Hypothesis (H0):** μ = 500 g
- **Alternative Hypothesis (Ha):** μ != 500 g

   This is a **two-tailed test** because we want to determine whether the average weight is different from 500 g, either higher or lower.

**Q6.** When should you use a Z-test instead of a t-test?

**Ans-** **Z-Test:** A Z-test is used when,

 1. The **population standard deviation (σ)** is **Known**.
 2. The **sample size** is large **(n >= 30)**

  **t-Test:** A t-test is used when,

 - The **population standard deviation (σ)** is **Unknown**.
 - The **sample size** is small **(n < 30)**.

| **Z-Test**                                        | **t-Test**                                          |
| ------------------------------------------------- | --------------------------------------------------- |
| Population standard deviation (σ) is known        | Population standard deviation is unknown                                |
| Z = (​Xˉ−μ​)/σ/n                                    | t = (Xˉ−μ​)/s/n |
| Large sample size (n >= 30)                       | Small sample size (n < 30)                                                 |
| Uses the standard normal (Z) distribution         | Uses the t-distribution                                      |
